```python
import trimesh
import numpy as np

mesh = trimesh.load('/content/meshes/Meshes/girl.obj')

vertices = mesh.vertices
print("Vertices:", vertices.shape[0])
for i, axis in enumerate(['X','Y','Z']):
    print(axis,
          "min:", np.min(vertices[:,i]),
          "max:", np.max(vertices[:,i]),
          "mean:", np.mean(vertices[:,i]),
          "std:", np.std(vertices[:,i]))

mesh.show()
```
